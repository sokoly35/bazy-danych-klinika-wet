# WŁAŚCICIELE

### Imiona właścicieli

```python
m_i = pd.read_csv('dane do tabeli/mezczyzni_imiona.csv')
k_i = pd.read_csv('dane do tabeli/kobiety_imiona.csv')
```

```python
liczba_m = np.random.rand(1)[0]
liczba_m = np.round(liczba_m*1000)
liczba_m = int(liczba_m)
liczba_k = 1000-liczba_m
liczba_m, liczba_k
```

```python
m_i_suma = sum(m_i["LICZBA_WYSTĄPIEŃ"])
m_i_suma
```

```python
k_i_suma = sum(k_i["LICZBA_WYSTĄPIEŃ"])
k_i_suma
```

```python
m_i["PRAWDOPODOBIEŃSTWO"] = m_i["LICZBA_WYSTĄPIEŃ"]/m_i_suma
k_i["PRAWDOPODOBIEŃSTWO"] = k_i["LICZBA_WYSTĄPIEŃ"]/k_i_suma
```

```python
m_i.head()
```

```python
k_i.head()
```

```python
m_imiona = m_i.sample(liczba_m, weights=m_i['PRAWDOPODOBIEŃSTWO'], replace=True, random_state=7)
m_imiona = m_imiona.reset_index()
m_imiona = m_imiona.drop(columns=['index','PŁEĆ'])
m_imiona
```

```python
k_imiona = k_i.sample(liczba_k, weights=k_i['PRAWDOPODOBIEŃSTWO'], replace=True, random_state=7)
k_imiona = k_imiona.reset_index()
k_imiona = k_imiona.drop(columns=['index','PŁEĆ'])
k_imiona
```

### Nazwiska właścicieli

```python
m_n = pd.read_csv('dane do tabeli/mezczyzni_nazwiska.csv')
k_n = pd.read_csv('dane do tabeli/kobiety_nazwiska.csv')
```

```python
m_n_suma = sum(m_n["Liczba"])
m_n_suma
```

```python
k_n_suma = sum(k_n["Liczba"])
k_n_suma
```

```python
m_n["PRAWDOPODOBIEŃSTWO"] = m_n["Liczba"]/m_n_suma
k_n["PRAWDOPODOBIEŃSTWO"] = k_n["Liczba"]/k_n_suma
```

```python
m_nazwiska = m_n.sample(liczba_m, weights=m_n['PRAWDOPODOBIEŃSTWO'], replace=True, random_state=7)
m_nazwiska = m_nazwiska.reset_index()
m_nazwiska = m_nazwiska.drop(columns=['index'])
m_nazwiska
```

```python
k_nazwiska = k_n.sample(liczba_k, weights=k_n['PRAWDOPODOBIEŃSTWO'], replace=True, random_state=7)
k_nazwiska = k_nazwiska.reset_index()
k_nazwiska = k_nazwiska.drop(columns=['index'])
k_nazwiska
```

## Imiona i nazwiska właścicieli

```python
m1 = m_imiona["IMIĘ_PIERWSZE"]
m2 = m_nazwiska["Nazwisko aktualne"]
mezczyzni = pd.concat([m1, m2], axis=1)
mezczyzni.columns = ['Imię', 'Nazwisko']
mezczyzni
```

```python
k1 = k_imiona["IMIĘ_PIERWSZE"]
k2 = k_nazwiska["Nazwisko aktualne"]
kobiety = pd.concat([k1, k2], axis=1)
kobiety.columns = ['Imię', 'Nazwisko']
kobiety
```

```python
osoby = pd.concat([mezczyzni, kobiety])
osoby = osoby.reset_index()
osoby = osoby.drop(columns=['index'])
osoby
```

```python
osoby['Imię'] = osoby['Imię'].str.capitalize() 
osoby['Nazwisko'] = osoby['Nazwisko'].str.capitalize() 
```

## PESEL

Daty od 01.01.1950 do 20.06.2003 (pełnoletni)

Poniżej korzystam z generatora online: https://pesel.cstudios.pl/O-generatorze/Generator-On-Line

```python
pesele = pd.read_csv('dane do tabeli/pesele.csv', dtype={'mężczyźni':str, 'kobiety':str})
pesele['mężczyźni'] = pesele['mężczyźni'].apply(lambda x: '0' + x if len(x) < 11 else x)
pesele['kobiety'] = pesele['kobiety'].apply(lambda x: '0' + x if len(x) < 11 else x)
# bo mogą być pesele zaczynające się na 00
pesele['mężczyźni'] = pesele['mężczyźni'].apply(lambda x: '0' + x if len(x) < 11 else x)
pesele['kobiety'] = pesele['kobiety'].apply(lambda x: '0' + x if len(x) < 11 else x)
pesele
```

```python
pesele
```

```python
pesel1 = pesele.loc[0:liczba_m-1, 'mężczyźni']
pesel2 = pesele.loc[0:liczba_k-1, 'kobiety']

pesele = pd.concat([pesel1, pesel2])
pesele = pd.DataFrame(pesele)
pesele.columns = ['PESEL']
pesele = pesele.reset_index()
pesele = pesele.drop(columns=['index'])
pesele
```

```python
osoby = pd.concat([osoby, pesele], axis=1)
```

## Emaile

```python
domeny = ["gmail.com", "wp.pl", "onet.pl", "o2.pl", "interia.pl", "yahoo.com", "hotmail.com"]
```

```python
osoby["random_numer"] = np.random.randint(0,1000,1000)
osoby["random_domena"] = np.random.randint(0,6,1000)
osoby["domena"] = osoby["random_domena"].apply(lambda x: domeny[x])
osoby["Email"] = osoby["Imię"].str.lower() + '.' + osoby["Nazwisko"].str.lower() + osoby['random_numer'].astype(str) + '@' + osoby['domena']
osoby = osoby.drop(columns=['random_numer','random_domena','domena'])
```

```python
osoby = osoby.reset_index()
osoby.rename(columns = {'index':'id_właściciela'},inplace=True)
```

```python
osoby.to_csv('właściciele.csv')
```
